```python
pd.concat([pd.DataFrame(pd.json_normalize(x)) for x in df_indicador.dc_formula_indicador.apply(json.loads)])
```

```python
formulas = []

for i, r in df_indicador.iterrows():
#     formulas.append(pd.json_normalize(json.loads(r.dc_formula_indicador)))
    df_j = pd.json_normalize(json.loads(r.dc_formula_indicador))
    df_j['cd_indicador'] = r.cd_indicador
    formulas.append(df_j)
```

```python
df_formulas = pd.concat(formulas)
```

```python
df_formulas[df_formulas.cd_indicador == 418.0]
```

```python
# indicador --> Formula --> Variavel --> Variável_resultado
```

```python
df_indicador_variavel_resultado = df_indicador.merge(df_formulas, how='left', on='cd_indicador')\
.merge(df_variavel, how='left', left_on='caractere', right_on='nm_resumido_variavel')\
.merge(df_variavel_resultado, how='left', on='cd_variavel')
```

```python
df_indicador_variavel_resultado.columns
```

```python
df_indicador_variavel_resultado.loc[~df_indicador_variavel_resultado.in_consideracao_regiao, \
                                    ['cd_indicador',\
                                    'nm_indicador',\
                                    'cd_periodo',\
                                    'cd_regiao']].\
groupby(by=['cd_indicador']).\
agg({
    'nm_indicador':'first',
#     'cd_periodo':'count',
    'cd_regiao':'nunique'
})#.to_csv('resultados/indicadores_e_contagem_variavel_por_regiao.csv')
```
